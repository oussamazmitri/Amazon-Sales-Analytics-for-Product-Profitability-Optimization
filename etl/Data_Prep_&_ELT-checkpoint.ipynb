standardize Status column

```python
def normalize_status(status):
    if status.startswith("Cancelled"):
        return "Cancelled"
    if status.startswith("Pending"):
        return "Pending"
    if "Delivered" in status:
        return "Delivered"
    if any(x in status for x in ["Returned", "Rejected", "Damaged", "Lost"]):
        return "Failed"
    if status.startswith("Shipped") or status == "Shipping":
        return "In Transit"
    return "Other"

new_data["Status Clean"] = new_data["Status"].apply(normalize_status)

print(new_data["Status Clean"].value_counts().sum())
```
